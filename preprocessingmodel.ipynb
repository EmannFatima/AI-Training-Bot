## Normalization

```python
game_play['x_coord_diff'] = game_play['player1_x_coord'] - game_play['player2_x_coord']
game_play['y_coord_diff'] = game_play['player1_y_coord'] - game_play['player2_y_coord']
```

```python
game_play.drop([
    'fight_result','has_round_started','is_round_over',
    'player1_x_coord', 'player1_y_coord', 'player2_x_coord', 'player2_y_coord','player1_move_id',
    'player2_move_id', 'player2_buttons_up', 'player2_buttons_down', 'player2_buttons_right', 'player2_buttons_left', 'player2_button_X',
    'player2_button_Y', 'player2_button_A', 'player2_button_B', 'player2_button_R', 'player2_button_L'], axis=1, inplace=True)
```

```python
# Select numeric columns
numeric_cols = game_play.select_dtypes(include=['int64']).columns
```

```python
# Fit and transform the selected columns using the scaler object
game_play[numeric_cols] = scaler.fit_transform(game_play[numeric_cols])
```

```python
game_play['health 1'] = game_play['player1_health']/100
game_play['health 2'] = game_play['player2_health']/100
```

```python
game_play[['player1_is_jumping','player1_is_crouching','player1_is_player_in_move', 'player1_buttons_up', 'player1_buttons_down', 'player1_buttons_right',
           'player1_buttons_left', 'player1_button_X', 'player1_button_Y', 'player1_button_A', 'player1_button_B', 'player1_button_R', 'player1_button_L',
           'player2_is_jumping', 'player2_is_crouching', 'player2_is_player_in_move']] = game_play[['player1_is_jumping','player1_is_crouching','player1_is_player_in_move', 
                                                                                                    'player1_buttons_up', 'player1_buttons_down', 'player1_buttons_right',
                                                                                                    'player1_buttons_left', 'player1_button_X', 'player1_button_Y', 'player1_button_A', 
                                                                                                    'player1_button_B', 'player1_button_R', 'player1_button_L',
                                                                                                    'player2_is_jumping', 'player2_is_crouching', 'player2_is_player_in_move']].astype('int64')
           
           
 
```

## Combining Cells

```python
def GetMovesPerformed(row):
    Move = ''.join([str(x) for x in row])
    return Move

MoveColumns = ['player1_buttons_up', 'player1_buttons_down', 'player1_buttons_right',
        'player1_buttons_left', 'player1_button_X', 'player1_button_Y', 'player1_button_A', 
        'player1_button_B', 'player1_button_R', 'player1_button_L',]

game_play['buttons1'] = game_play[MoveColumns].apply(GetMovesPerformed, axis=1)

game_play.drop(MoveColumns, axis=1, inplace=True)
```

```python
game_play[['buttons1']] = game_play[['buttons1']].astype('bytes')
game_play['buttons_1'] = game_play['buttons1'].apply(lambda x: int(x, 2))

# #Drop the button1 Column
game_play.drop(['buttons1'], axis=1, inplace=True)
```
